```python
frames = [citi_bike_Jun2018_tripdata, citi_bike_Jul2018_tripdata, citi_bike_Aug2018_tripdata, 
          citi_bike_Sep2018_tripdata, citi_bike_Oct2018_tripdata, citi_bike_Nov2018_tripdata, 
          citi_bike_Dec2018_tripdata, citi_bike_Jan2019_tripdata, citi_bike_Feb2019_tripdata, 
          citi_bike_Mar2019_tripdata, citi_bike_Apr2019_tripdata, citi_bike_May2019_tripdata]

yearFive_df = pd.concat(frames)
yearFive_df
```

```python
yearFive_rename_df = yearFive_df["starttime"].str.split(" ", n=1, expand=True)
yearFive_rename_df = yearFive_rename_df.rename(columns={0: "Start date", 1: "Start time"})
yearFive_rename_df
```

```python
gender_df = yearFive_df["gender"].replace([0, 1, 2], ["Undefined", "Male", "Female"]) 
gender_df
```

```python
yearFive_final_df = yearFive_df[["tripduration", "start station id", "start station name", "start station latitude", 
                               "start station longitude", "end station id", "end station name", "end station latitude",
                              "end station longitude", "bikeid", "usertype", "birth year"]]
yearFive_final_df
```

```python
yearFive_final_df["Start Date"] = yearFive_rename_df ["Start date"]
yearFive_final_df["Start Time"] = yearFive_rename_df ["Start time"]
yearFive_final_df["Gender"] = gender_df
yearFive_final_df
```
