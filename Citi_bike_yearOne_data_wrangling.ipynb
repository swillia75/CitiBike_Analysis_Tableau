```python
yearOne_rename_df = yearOne_df["starttime"].str.split(" ", n=1, expand=True)
yearOne_rename_df = yearOne_rename_df.rename(columns={0: "Start date", 1: "Start time"})
yearOne_rename_df                                                     

```

```python
gender_df = yearOne_df["gender"].replace([0, 1, 2], ["undefined", "male", "female"]) 
gender_df
```

```python
yearOne_final_df["Start Date"] = yearOne_rename_df ["Start date"]
yearOne_final_df["Start Time"] = yearOne_rename_df ["Start time"]
yearOne_final_df["Gender"] = gender_df
yearOne_final_df
```
